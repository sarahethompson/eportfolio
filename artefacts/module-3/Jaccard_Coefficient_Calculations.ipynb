```python
from sklearn.metrics import confusion_matrix
import numpy as np

# Use the same encoded data for each person
# jack_data, mary_data, and jim_data were already created in a previous cell

def simple_matching_coefficient(y_true, y_pred):
    """Calculates the Simple Matching Coefficient for binary data."""
    # Ensure inputs are numpy arrays
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    # Calculate the confusion matrix components
    # C[0,0] is TN (0-0 matches)
    # C[0,1] is FP (0-1 mismatches)
    # C[1,0] is FN (1-0 mismatches)
    # C[1,1] is TP (1-1 matches)
    C = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = C.ravel()

    # Calculate SMC
    smc = (tp + tn) / (tp + tn + fp + fn)
    return smc

# Calculate Simple Matching Coefficient
print("Jack & Mary (SMC):", simple_matching_coefficient(jack_data, mary_data))
print("Jack & Jim (SMC):", simple_matching_coefficient(jack_data, jim_data))
print("Jim & Mary (SMC):", simple_matching_coefficient(jim_data, mary_data))
```

## Notes on Jaccard Similarity vs. Simple Matching Coefficient

This notebook explores calculating similarity between individuals based on their characteristics and test results using two different metrics: Jaccard Similarity and the Simple Matching Coefficient (SMC).

**Data Encoding:**
Before calculating similarity, categorical data (like Gender, Fever, Cough, and Test results) was encoded into a binary format (0s and 1s).

*   For columns with two categories (Gender, Fever, Cough), one category was assigned 1 and the other 0. While the assignment of 0 and 1 is mathematically arbitrary, assigning 1 to the "present" or "specific" characteristic (Male, Fever 'Y', Cough 'P') helps in interpreting similarity based on these attributes.
*   For the 'Test' columns, values 'Y', 'P', and 'A' were encoded as 1, and 'N' was encoded as 0. This encoding decision was based on the context of pathological tests, where 'Y', 'P', and 'A' likely represent abnormal or positive findings, and 'N' represents a normal or negative result. Encoding '1' for these non-'N' results allows us to measure similarity based on shared *abnormalities* or *present* findings.

**Jaccard Similarity:**
The Jaccard similarity coefficient is calculated as the size of the intersection of two sets divided by the size of their union. In the context of binary data, it specifically measures the overlap in **shared '1' values (shared presences)**. It **ignores shared '0' values (shared absences)**.

*   We initially calculated Jaccard similarity based on the encoded Test columns (`Test-1` to `Test-4`) and later included Gender, Fever, and Cough.
*   The Jaccard results showed the degree to which individuals shared positive or present characteristics based on the defined encoding.

**Simple Matching Coefficient (SMC):**
The Simple Matching Coefficient is a different similarity measure that considers **all matches**, both shared '1' values (presences) and shared '0' values (absences). It is calculated as the total number of matches (1-1 and 0-0) divided by the total number of attributes.

*   We calculated SMC because the user questioned why shared '0' values (like two individuals both having a 'Negative' test result) shouldn't contribute to overall similarity, especially when the inherent meaning of the categories isn't explicitly provided.
*   SMC aligns with the idea that having the *same* value, regardless of whether it's encoded as 0 or 1, contributes equally to the similarity score.

**Key Takeaway:**
The choice between Jaccard Similarity and the Simple Matching Coefficient depends on what aspects of similarity you want to emphasize.

*   Use **Jaccard Similarity** when you are primarily interested in the overlap of **shared present characteristics (1s)** and shared absences (0s) are not considered informative for similarity.
*   Use **Simple Matching Coefficient** when you believe **all shared attributes**, whether present (1s) or absent (0s), contribute equally to the overall similarity.

The different results obtained from Jaccard and SMC highlight the importance of understanding how each metric defines and measures similarity and choosing the one that best fits the analytical goal and the nature of the data.
