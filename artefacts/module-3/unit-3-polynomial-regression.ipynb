## Summary of Analysis

This notebook explores polynomial regression using a dataset of cars passing a tollbooth, recording their time of day (`x`) and speed (`y`).

1.  **Initial Exploration**: We started with a basic polynomial regression model and visualized the relationship between the original `x` and `y` data.

2.  **Assessing the Relationship**: We calculated the R-squared value to quantify how well the polynomial model fits the data. An R-squared value close to 1 indicates a strong relationship.

3.  **Predicting Future Values**: We used the polynomial model to predict the speed of a car at a specific time (17 P.M.).

4.  **Comparing Original and Modified Data**: To understand how changes in data points affect the regression, we added a visualization that plots both the original data and a modified set of data points. This allows for a visual comparison of the data distribution.

5.  **Combined Analysis Cell**: We consolidated the code for plotting both datasets, performing polynomial regression on the current data, and calculating the R-squared value into a single cell for easier experimentation.

By modifying the `x` and `y` values in the "Current Data" section of the final code cell, you can observe how the changes impact the scatter plot, the polynomial regression line, and the R-squared value, demonstrating the sensitivity of the model to data variations.

```python
import numpy
import matplotlib.pyplot as plt

# Original x: [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
# Original y: [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,95] # Modified a value here

#NumPy has a method that lets us make a polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

#specify how the line will display, we start at position 1, and end at position 22
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

It is important to know how well the relationship between the values of the x- and y-axis is, if there are no relationship the polynomial regression can not be used to predict anything.

The relationship is measured with a value called the r-squared.

The r-squared value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related

```python
import numpy
from sklearn.metrics import r2_score

# Original x: [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
# Original y: [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,95] # Modified a value here

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

## Predict Future Values

Let us try to predict the speed of a car that passes the tollbooth at around 17 P.M

```python
import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

speed = mymodel(17)
print(speed)
```

```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Original x: [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
original_x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
# Original y: [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
original_y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# Current x and y data (you can modify these to experiment)
x = [1,3,4,6,7,9,10,11,13,14,15,17,18,20,21,22,23,24] # Changed a few x values
y = [110,85,75,65,60,50,55,60,70,72,77,80,85,95,105,100,90,88] # Changed a few y values

# Plot original and current data points
plt.scatter(original_x, original_y, color='blue', label='Original Data')
plt.scatter(x, y, color='red', label='Current Data')
plt.xlabel("Time of Day (x)")
plt.ylabel("Speed (y)")
plt.title("Original vs. Current Data Points")
plt.legend()
plt.show()

# Polynomial regression
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

# Specify how the line will display
myline = numpy.linspace(1, 24, 100) # Adjusted line range based on new x values

# Plot the polynomial regression line for the current data
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

# Calculate and print R-squared value for the current data
print("R-squared value (Current Data):", r2_score(y, mymodel(x)))
```
