# Removing Missing Values from both datasets

When calculating a correlation between two variables, the calculation typically involves multiplying or summing values from both variables for each observation. If an observation has a missing value in either of the variables we are trying to correlate, that observation cannot be included in the calculation.

Most statistical software and libraries (like pandas in Python) will automatically exclude rows with missing values when computing correlations. However, explicitly removing missing values beforehand gives you more control and transparency over which data is being used. It also helps avoid potential errors or unexpected results that can occur when working with functions that might not handle missing values gracefully.

Essentially, to get an accurate correlation coefficient, we need complete pairs of data for the variables we are analyzing. Removing missing values ensures that our correlation calculation is based only on these complete pairs.



```python
dfpop_cleaned = dfpop.dropna()
dfgdp_cleaned = dfgdp.dropna()

print("Shape of dfpop after dropping missing values:", dfpop_cleaned.shape)
print("Shape of dfgdp after dropping missing values:", dfgdp_cleaned.shape)
```

# Task
Investigate any correlation between the mean population of each country and the mean per capita GDP (from 2001 to 2021) using the datasets "world_population.csv" and "gdp_csv.csv".

## Calculate per capita gdp

### Subtask:
Divide the GDP by the population for each country and year in the cleaned dataframes to get per capita GDP.


**Reasoning**:
Identify the year columns in both dataframes, align the dataframes by country code, and then calculate the per capita GDP for each year by dividing GDP by population. Create a new dataframe to store the results.



```python
# Identify year columns (assuming year columns are numeric strings from 1960 to 2021)
year_cols_pop = [col for col in dfpop_cleaned.columns if col.isdigit() and 1960 <= int(col) <= 2021]
year_cols_gdp = [col for col in dfgdp_cleaned.columns if col.isdigit() and 1960 <= int(col) <= 2021]

# Find common years
common_years = list(set(year_cols_pop) & set(year_cols_gdp))
common_years.sort()

# Select only common year columns and country information from both dataframes
dfpop_years = dfpop_cleaned[['Country Name', 'Country Code'] + common_years].copy()
dfgdp_years = dfgdp_cleaned[['Country Name', 'Country Code'] + common_years].copy()

# Convert year columns to numeric, coercing errors to NaN
for year in common_years:
    dfpop_years[year] = pd.to_numeric(dfpop_years[year], errors='coerce')
    dfgdp_years[year] = pd.to_numeric(dfgdp_years[year], errors='coerce')

# Merge dataframes on 'Country Code' to ensure country alignment
merged_df = pd.merge(dfgdp_years, dfpop_years, on='Country Code', suffixes=('_gdp', '_pop'))

# Calculate per capita GDP for each year
df_per_capita_gdp = pd.DataFrame()
df_per_capita_gdp['Country Name'] = merged_df['Country Name_gdp']
df_per_capita_gdp['Country Code'] = merged_df['Country Code']

for year in common_years:
    gdp_col = year + '_gdp'
    pop_col = year + '_pop'
    # Handle potential division by zero or NaN in population
    df_per_capita_gdp[year] = merged_df[gdp_col] / merged_df[pop_col]

# Display the first few rows of the resulting dataframe
display(df_per_capita_gdp.head())
```

## Filter by year

### Subtask:
Filter both the population and per capita GDP dataframes to include only the years from 2001 to 2021.


**Reasoning**:
Filter the dataframes to include only the years from 2001 to 2021 and the country information.



**Reasoning**:
The error indicates that the year '2021' is not present in the columns of the `dfpop_years` or `df_per_capita_gdp` dataframes. I need to inspect the columns of these dataframes to identify the correct range of available years and then filter accordingly.



```python
print("Columns in dfpop_years:", dfpop_years.columns.tolist())
print("Columns in df_per_capita_gdp:", df_per_capita_gdp.columns.tolist())
```

**Reasoning**:
The columns show that the data for years go up to 2020, not 2021. I will adjust the year range accordingly and filter the dataframes.



```python
year_cols_2001_2020 = [str(year) for year in range(2001, 2021)]

dfpop_filtered = dfpop_years[['Country Name', 'Country Code'] + year_cols_2001_2020].copy()
df_per_capita_gdp_filtered = df_per_capita_gdp[['Country Name', 'Country Code'] + year_cols_2001_2020].copy()

display(dfpop_filtered.head())
display(df_per_capita_gdp_filtered.head())
```

## Calculate mean values

### Subtask:
Calculate the mean population and mean per capita GDP for each country across the filtered years.


**Reasoning**:
Calculate the mean population and mean per capita GDP for each country across the filtered years (2001-2020).



```python
year_cols_filtered_pop = [col for col in dfpop_filtered.columns if col.isdigit()]
dfpop_filtered['mean_population'] = dfpop_filtered[year_cols_filtered_pop].mean(axis=1)

year_cols_filtered_gdp = [col for col in df_per_capita_gdp_filtered.columns if col.isdigit()]
df_per_capita_gdp_filtered['mean_per_capita_gdp'] = df_per_capita_gdp_filtered[year_cols_filtered_gdp].mean(axis=1)

display(dfpop_filtered.head())
display(df_per_capita_gdp_filtered.head())
```

## Merge dataframes

### Subtask:
Merge the dataframes containing the mean population and mean per capita GDP on the 'Country Code' column.


**Reasoning**:
Select the required columns from the filtered dataframes and merge them on 'Country Code', then display the head of the merged dataframe.



```python
dfpop_means = dfpop_filtered[['Country Code', 'mean_population']]
dfgdp_means = df_per_capita_gdp_filtered[['Country Code', 'mean_per_capita_gdp']]

df_merged_means = pd.merge(dfpop_means, dfgdp_means, on='Country Code')

display(df_merged_means.head())
```

## Calculate correlation

### Subtask:
Calculate the correlation coefficient between the mean population and mean per capita GDP.


**Reasoning**:
Calculate the correlation matrix for the merged dataframe and extract the correlation coefficient between the mean population and mean per capita GDP, then print the coefficient.



```python
correlation_matrix = df_merged_means[['mean_population', 'mean_per_capita_gdp']].corr()
correlation_coefficient = correlation_matrix.loc['mean_population', 'mean_per_capita_gdp']
print(f"The correlation coefficient between mean population and mean per capita GDP is: {correlation_coefficient}")
```

## Visualize correlation

### Subtask:
(Optional) Create a scatter plot to visualize the relationship between mean population and mean per capita GDP.


**Reasoning**:
Create a scatter plot to visualize the relationship between mean population and mean per capita GDP using the merged dataframe.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df_merged_means['mean_population'], df_merged_means['mean_per_capita_gdp'])
plt.title('Scatter Plot of Mean Population vs. Mean Per Capita GDP (2001-2020)')
plt.xlabel('Mean Population')
plt.ylabel('Mean Per Capita GDP')
plt.grid(True)
plt.show()
```

## Perform Linear Regression

### Subtask:
Perform linear regression analysis with mean population as the independent variable and mean per capita GDP as the dependent variable.

**Reasoning**:
Use `statsmodels.api` to perform ordinary least squares (OLS) regression. Define the independent and dependent variables, add a constant to the independent variable for the intercept, fit the model, and display the regression summary.

```python
import statsmodels.api as sm

# Define the independent variable (X) and the dependent variable (y)
X = df_merged_means['mean_population']
y = df_merged_means['mean_per_capita_gdp']

# Combine X and y into a single DataFrame to easily drop rows with NaNs in either column
data_for_regression = pd.DataFrame({'mean_population': X, 'mean_per_capita_gdp': y})

# Drop rows with NaN values in either column
data_for_regression_cleaned = data_for_regression.dropna()

# Redefine X and y from the cleaned DataFrame
X_cleaned = data_for_regression_cleaned['mean_population']
y_cleaned = data_for_regression_cleaned['mean_per_capita_gdp']


# Add a constant to the independent variable for the intercept
X_cleaned = sm.add_constant(X_cleaned)

# Create and fit the OLS model
model = sm.OLS(y_cleaned, X_cleaned).fit()

# Print the regression summary
print(model.summary())
```

## Perform Linear Regression using Scikit-learn

### Subtask:
Perform linear regression analysis using `scikit-learn` with mean population as the independent variable and mean per capita GDP as the dependent variable.

**Reasoning**:
Import the `LinearRegression` model from `sklearn.linear_model`. Reshape the independent variable to be a 2D array as required by `sklearn` models. Create and fit the `LinearRegression` model, and then print the coefficient and intercept.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd # Import pandas

# Define the independent variable (X) and the dependent variable (y)
X = df_merged_means['mean_population']
y = df_merged_means['mean_per_capita_gdp']

# Combine X and y into a single DataFrame to easily drop rows with NaNs in either column
data_for_regression_sklearn = pd.DataFrame({'mean_population': X, 'mean_per_capita_gdp': y})

# Drop rows with NaN values in either column
data_for_regression_sklearn_cleaned = data_for_regression_sklearn.dropna()

# Redefine X and y from the cleaned DataFrame and reshape X for sklearn
X_cleaned_sklearn = data_for_regression_sklearn_cleaned['mean_population'].values.reshape(-1, 1)
y_cleaned_sklearn = data_for_regression_sklearn_cleaned['mean_per_capita_gdp'].values


# Create and fit the Linear Regression model
model_sklearn = LinearRegression()
model_sklearn.fit(X_cleaned_sklearn, y_cleaned_sklearn)

# Print the coefficient and intercept
print(f"Coefficient: {model_sklearn.coef_[0]:.4f}")
print(f"Intercept: {model_sklearn.intercept_:.4f}")
```

## Visualize Linear Regression

### Subtask:
Add the linear regression line from the scikit-learn model to the scatter plot of mean population vs. mean per capita GDP.

**Reasoning**:
Use the `matplotlib.pyplot` library to create a scatter plot of the cleaned mean population and mean per capita GDP data. Then, use the fitted `sklearn` model to predict the per capita GDP values based on the mean population and plot the resulting regression line on the scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Use the cleaned data from the sklearn regression
X_plot = data_for_regression_sklearn_cleaned['mean_population']
y_plot = data_for_regression_sklearn_cleaned['mean_per_capita_gdp']

# Reshape X_plot for prediction
X_plot_reshaped = X_plot.values.reshape(-1, 1)

# Predict the per capita GDP values using the fitted sklearn model
y_pred = model_sklearn.predict(X_plot_reshaped)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X_plot, y_plot, label='Data Points')

# Plot the regression line
plt.plot(X_plot, y_pred, color='red', linewidth=2, label='Linear Regression Line')

plt.title('Scatter Plot of Mean Population vs. Mean Per Capita GDP with Regression Line (2001-2020)')
plt.xlabel('Mean Population')
plt.ylabel('Mean Per Capita GDP')
plt.grid(True)
plt.legend()
plt.show()
```
