In general, all runs showed a strong positive Pearson correlation, indicating that as Data 1 increases, Data 2 tends to increase as well. The covariance values also support this positive relationship. The exact correlation values varied slightly depending on the number of data points and the specific parameters used to generate the data.

```python
# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr

# Function to generate data and calculate correlation and regression
def analyze_correlation_and_regression(num_points, data1_mean, data1_std, data2_mean_offset, data2_std):
    # seed random number generator
    seed(1)

    # prepare data
    data1 = data1_std * randn(num_points) + data1_mean
    data2 = data1 + (data2_std * randn(num_points) + data2_mean_offset)

    # calculate covariance matrix
    covariance = cov(data1, data2)

    # calculate Pearson's correlation
    corr, _ = pearsonr(data1, data2)

    # plot
    plt.scatter(data1, data2)
    plt.title(f'Scatter Plot with {num_points} points')
    plt.xlabel('Data 1')
    plt.ylabel('Data 2')
    plt.show()

    # summarize
    print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
    print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
    print('Covariance: %.3f' % covariance[0][1])
    print('Pearsons correlation: %.3f' % corr)
    print("-" * 30) # Separator for multiple runs

# Analyze with initial parameters
analyze_correlation_and_regression(1000, 100, 20, 50, 10)

# You can call the function again with different parameters to see how they affect the results
analyze_correlation_and_regression(500, 50, 10, 20, 5)
analyze_correlation_and_regression(2000, 200, 30, 100, 20)
```
